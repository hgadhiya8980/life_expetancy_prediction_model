```python
X = df.drop("Life expectancy", axis=1)
y = df["Life expectancy"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
lr.score(X_test, y_test)
```

```python
rd = Ridge()
rd.fit(X_train, y_train)
y_pred1 = rd.predict(X_test)
rd.score(X_test, y_test)
```

```python
lc = Lasso()
lc.fit(X_train, y_train)
y_pred2 = lc.predict(X_test)
lc.score(X_test, y_test)
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)
y_pred3 = svr.predict(X_test)
svr.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
y_pred4 = dtr.predict(X_test)
dtr.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
y_pred5 = rfr.predict(X_test)
rfr.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cv = cross_val_score(rfr, X_train, y_train, cv=10)
cv, cv.mean()
```

```python
def life_insurance_prediction(model,entity, Year):
    for num,var in enumerate(df_value):
        if var == entity:
            entity = num
    x = np.zeros(len(X.columns))
    x[0] = entity
    x[1] = Year
    
    x = sc.transform([x])[0]
    return model.predict([x])[0]
```
